# Review of images

Images are 2D arrays or Matrices, they usually have the following form: (Height, Width, Channels), Usually Images have 3 channels which are RGB. But, they can have more channels depending on applications (e.g. Biomedical images).

<img src = "frames/gait/1.jpg">

Here you have some special details that you need to take into account:

+ OpenCV reads BGR instead of RGB, so you need to flip the order using ::-1 at the final index.
+ Since Images are matrices, you can perform some mathematical operations and we are going to do some of them in the next exercises.
+ Remember to make operations using np.float64, and np.uint8 just for visualization.

```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
from ipywidgets import interact, interact_manual, fixed
import ipywidgets as widgets

def plot_images(path):
    image = cv.imread(path)
    plt.figure(figsize=(8,8))
    plt.title("image")
    plt.grid()
    plt.xlabel("width - cols")
    plt.ylabel("height - rows")
    plt.imshow(image)
```

```python
path = "frames/gait/1.jpg"
plot_images(path)
```

# Review of videos

Videos can be seen like a list of frames that are stacked, so a video will have the following form: (height, width, channels, frames). 

<img src="../imgs/video_frames.PNG">

in the following exercises, consider channels = 1 (if you have more than 1 channel, you can obtain the mean of the channels).

```python
def get_video(path):
    print("The video path is the following: {}".format(path))
    video = cv.VideoCapture(path)
    height = int(video.get(cv.CAP_PROP_FRAME_HEIGHT))
    width = int(video.get(cv.CAP_PROP_FRAME_WIDTH))
    frames = int(video.get(cv.CAP_PROP_FRAME_COUNT))
    channels = 1
    print ("The height of the video is: {} pixels".format(height))
    print ("The width of the video is: {} pixels".format(width))
    print ("The video has {} frames ".format(frames))
    print ("Each frame of the video has {} channels".format(channels))
    has_frame,current_frame = video.read()
    video_list = []
    while has_frame :
        current_frame = cv.cvtColor(current_frame, cv.COLOR_BGR2GRAY)
        video_list.append(current_frame)
        has_frame, current_frame = video.read()
    return video_list

def plot_a_frame(image, frame):
    image = image[frame]
    image = np.array(image,np.uint8)
    plt.imshow(image, cmap = "gray")
```

```python
path = "../videos/gait.mp4"
video = get_video(path)

interact(plot_a_frame, frame=widgets.IntSlider(min=0,max=len(video)-1, step=1,value=10), 
        image = fixed(video))
```
